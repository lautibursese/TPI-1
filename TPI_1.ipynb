+ Máxima duración según tipo de film (película/serie), por plataforma y por año:
    El request debe ser: get_max_duration(año, plataforma, [min o season])

```python
def get_max_duration(año=int, plataforma=str, min_o_season=str):
    mascara1 = df_definitivo['Release_Year'] == año
    mascara2 = df_definitivo['Platform'] == plataforma
    mascara3 = df_definitivo['Duration'].max()
    mascara_movie = df_definitivo['Type'] == 'Movie'
    mascara_season = df_definitivo['Type'] == 'TV Show'
    if type(año) != int or type(plataforma) != str or type(min_o_season) != str:
        print('Please, enter an integer year and the two other parameters in string type')
    elif año not in (df_definitivo['Release_Year'].unique()):
        print('Please select one of these years', df_definitivo['Release_Year'].unique())
    elif plataforma not in (df_definitivo['Platform'].unique()):
        print('Please select one of these platforms', df_definitivo['Platform'].unique())
    elif min_o_season not in ('min', 'season'):
        print('Please enter "min" or "season"')
    elif min_o_season == 'season':
        print(df_definitivo[mascara1&mascara2&mascara_season].sort_values('Duration')['Duration'].tail(1))
    else:
        return df_definitivo[mascara1&mascara2&mascara_movie].sort_values('Duration')['Duration'].tail(1)

get_max_duration(2018, 'Hulu', 'min')
```

+ Cantidad de películas y series (separado) por plataforma
    El request debe ser: get_count_plataform(plataforma)  

```python
def get_count_platform(plataforma):
    mascaro1 = df_definitivo['Platform'] == plataforma
    mascaro_peliculas = df_definitivo['Type'] == 'Movie'
    mascaro_seasons = df_definitivo['Type'] == 'TV Show'
    if type(plataforma) != str:
        print('Please enter a string value: "Amazon_Prime", "Disney_Plus", "Hulu" or "Netflix"')
    else:
        return print(df_definitivo[mascaro1&mascaro_peliculas].count()[0].tolist(), 'peliculas y', df_definitivo[mascaro1&mascaro_seasons].count()[0], 'series')

get_count_platform('Netflix')
```

+ Cantidad de veces que se repite un género y plataforma con mayor frecuencia del mismo.
    El request debe ser: get_listedin('genero')  
    Como ejemplo de género pueden usar 'comedy', el cuál deberia devolverles un cunt de 2099 para la plataforma de amazon.

```python
def get_listedin(genero):
    mas = df_definitivo['Listed_in'].str.contains(genero, case=False)
    mas_amazonprime = df_definitivo['Platform'] == 'Amazon_Prime'
    mas_disneyplus = df_definitivo['Platform'] == 'Disney_Plus'
    mas_hulu = df_definitivo['Platform'] == 'Hulu'
    mas_netflix = df_definitivo['Platform'] == 'Netflix'
    if type(genero) != str:
        print('Please enter a string value')
    else:
        return f"Amazon Prime cuenta con {df_definitivo[mas&mas_amazonprime].count()[0]}, Disney PLus {df_definitivo[mas&mas_disneyplus].count()[0]}, Hulu {df_definitivo[mas&mas_hulu].count()[0]} y Netflix {df_definitivo[mas&mas_netflix].count()[0]}"
    
get_listedin('comedy')
```

+ Actor que más se repite según plataforma y año.
  El request debe ser: get_actor(plataforma, año)

```python
#ACOTACION IMPORTANTE, HABIAN MUCHAS COSAS SIN DATO, Y REALMENTE NO LOS HABIA, SE TUVO QUE LIMITAR LA FUNCION A QUE EMITA 2 ACTORES
#Y NO UNO, YA QUE MUCHAS VECES, EL MAYOR REPETIDO ES "SIN DATO". A VER SI SE TIENEN MAS COMPLETAS LAS BASES... JAJA

from collections import Counter
import itertools

def get_actor(plataforma, año):
    prueba = itertools.chain(*df_definitivo['Cast'].str.split(','))
    hola = Counter(list(prueba)).most_common()[1]
    mas_platform = df_definitivo['Platform'] == plataforma
    mas_year = df_definitivo['Release_Year'] == año
    prueba2 = list(itertools.chain(*df_definitivo[mas_platform&mas_year]['Cast'].str.split(',')))
    if type(plataforma) != str:
        print(f'Please enter a string value in platform parameter')
    elif plataforma not in (df_definitivo['Platform'].unique()):
        print(f"Please enter one of these {df_definitivo['Platform'].unique()}")
    elif type(año) != int:
        print('Please enter an int value in year parameter')
    else:
        return f"Los dos actores que mas se repiten son {Counter(prueba2).most_common()[0:2]}"

get_actor('Netflix', 2018)
```
